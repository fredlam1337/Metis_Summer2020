```python
import pandas as pd
import numpy as np
import ast
```

### Choose which dataset to clean

```python
# df = pd.read_csv('TFT_Platinum_MatchData.csv')
# df = pd.read_csv('TFT_Diamond_MatchData.csv')
# df = pd.read_csv('TFT_Master_MatchData.csv')
# df = pd.read_csv('TFT_GrandMaster_MatchData.csv')
df = pd.read_csv('TFT_Challenger_MatchData.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
comb_str = df['combination']
res = []

for i in comb_str:
    a = ast.literal_eval(i)
    res.append(a)
```

```python
com_df = pd.DataFrame(res)
```

```python
com_df.fillna(0, inplace=True)
com_df = com_df.astype(int)
com_df
```

```python
com_df = com_df.reindex(sorted(com_df.columns), axis=1)
```

```python
champ_str = df['champion']
res_3 = []

for i in champ_str:
    res_2 = {}
    b = ast.literal_eval(i)
    for j in b:
        res_2[j + '_items'] = len(b[j]['items'])
        res_2[j + '_star'] = b[j]['star']
    res_3.append(res_2)
```

```python
champ_df = pd.DataFrame(res_3)
```

```python
champ_df.fillna(0, inplace=True)
champ_df = champ_df.astype(int)
champ_df
```

```python
champ_df = champ_df.reindex(sorted(champ_df.columns), axis=1)
```

```python
f_df = pd.concat([df, com_df, champ_df], axis=1, join='inner')
f_df.drop(columns = ['combination', 'champion'], inplace=True)
```

```python
f_df.replace({'Ranked' : { 8 : 'Lose', 7 : 'Lose', 6 : 'Lose', 5: 'Lose', 4: 'Lose', 3: 'Win', 2: 'Win', 1: 'Win' }}, inplace=True)
```

```python
f_df = f_df[f_df['Ranked'] != 0]
```
