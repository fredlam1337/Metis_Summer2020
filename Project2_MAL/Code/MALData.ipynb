### Data

```python
#read in csv with columns: links and anime number
#you can scrape to create the csv but many shows have missing values so you won't get as much data as you'd like
#so either pull a lot of shows or set up your own csv
link_df = pd.read_csv('anime_links.csv')
link_df.head(5)
```

```python
#Disclaimer: THIS TAKES A WHILE
ani_list = []

for index, row in link_df.iterrows():
    a = ani_details(row['link'])
    b = get_episodes(row['link'])
#Doesn't work
#     c = get_characters(row['link'])
    d = get_clubs(row['link'])
#Doesn't work
#     e = get_reviews(row['link'])
    f = forum_posts(str(row['number']))
    g = forum_replies(str(row['number']))
    
    ani_dict = {**a, **b, **d, **f, **g}
#Genre creates too many dummy features
#     for genre in ani_dict['genres']:
#         ani_dict[genre] = 'Yes'

#     del ani_dict['genres']
    ani_list.append(ani_dict)
    
#     print(row['link'], row['number'])
```

```python
ani_list
```

```python
#set list to pandas dataframe and save to csv
ani_df = pd.DataFrame(ani_list)
ani_df.set_index('title', inplace = True)

ani_df.to_csv('ani_data.csv')
```
